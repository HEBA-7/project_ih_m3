### Cruce de tablas

```python
#concateno las tres tablas últimas que tienen el mismo número de rows 

df_concat = pd.concat([di_trans, di_dime, di_prop, ], ignore_index=False, axis=1)
```

```python
df_concat
```

```python
#ahora toca juntar el resto de las tablas para poder empezar mergearlas con la tabla creada arriba df_concat(la que hemos concatenado)
```

```python
#jutnto la de las ciudades con la df_concat

df_join1 = [df_concat, di_city]
df_city = ft.reduce(lambda left, right: pd.merge(left, right, on='city_id'), df_join1)
df_city
```

```python
#junto la de la df_city con la de clarity


df_join2 = [df_city, di_clarity]
df_clarity = ft.reduce(lambda left, right: pd.merge(left, right, on='clarity_id'), df_join2)
df_clarity
```

```python
#junto la df_clarity con la de color

df_join3 = [df_clarity, di_color]
df_color = ft.reduce(lambda left, right: pd.merge(left, right, on='color_id'), df_join3)
df_color
```

```python
# y finalmente junto la df_color con la de cut, con esta acabaríamos de juntarlas todas

df_join4 = [df_color, di_cut]
df_cut = ft.reduce(lambda left, right: pd.merge(left, right, on='cut_id'), df_join4)
df_cut
```

```python
df_cut.info()
```

```python
#quitamos las columnas duplicadas como es el index_id

df = df_cut.T.drop_duplicates().T
df
```
